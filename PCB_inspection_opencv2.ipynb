```
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

читаем изображения, показываем их

```
img1 = cv2.imread('pcb_original.png', cv2.IMREAD_COLOR)
img2 = cv2.imread('pcb_test.png', cv2.IMREAD_COLOR)
cv2_imshow(img1)
img1.shape
img2.shape
cv2_imshow(img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

размер ячеек


```
print(img1.shape)
print(img2.shape)
```

главный код. сначала конвертируем в серый, потом в бинарный формат (черно белый) потом XOR операция

```
gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
ret, thresh1 = cv2.threshold(gray1, 200, 255, cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(gray2, 200, 255, cv2.THRESH_BINARY)
res = cv2.bitwise_xor(thresh1, thresh2, mask=None)
cv2_imshow(res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
